# Instancias aleatorias y sus soluciones óptimas

```python
# -*- coding: utf-8 -*-

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('ggplot')
```

## Lectura de archivos

```python
namesHead4 = ['Tiempo', 'Gyms', 'Pokeparadas', 'Mochila', 'Distancia','Estaciones visitadas', 'Nro. Ej', 'Vecindad', 'RCL', 'Limite']
namesHead3 = ['Tiempo', 'Gyms', 'Pokeparadas', 'Mochila', 'Distancia','Estaciones visitadas', 'Nro. Ej', 'Vecindad']
namesHead2 = ['Tiempo', 'Gyms', 'Pokeparadas', 'Mochila', 'Distancia','Estaciones visitadas', 'Nro. Ej']
namesHead = ['Tiempo', 'Gyms', 'Pokeparadas', 'Mochila', 'Distancia','Estaciones visitadas', 'Nro. Ej']
ejemplo = pd.read_csv('salidaEj5/random_16_inputs.csv', header = None, names = namesHead)
ejemplo2 = pd.read_csv('salidaEj5/random_16_inputs_ej2.csv', header = None, names = namesHead2)
ejemplo3 = pd.read_csv('salidaEj5/random_16_inputs_ej3.csv', header = None, names = namesHead3)
ejemplo4 = pd.read_csv('salidaEj5/random_16_inputs_ej4.csv', header = None, names = namesHead4)
ejemplo99_2 = pd.read_csv('salidaEj5/random_99_inputs.csv', header = None, names = namesHead2)
ejemplo99_3 = pd.read_csv('salidaEj5/random_99_inputs_ej3.csv', header = None, names = namesHead3)
ejemplo99_4 = pd.read_csv('salidaEj5/random_99_inputs_ej4.csv', header = None, names = namesHead4)
```

## Tiempo vs cantidad de nodos

```python

config1 = ejemplo
#config1 = config1.groupby('RCL')
config2 = ejemplo2
#config2 = config2.groupby('RCL')
config3 = ejemplo3
#config3 = config3.groupby('RCL')
config4 = ejemplo4
#config4 = config4.groupby('RCL')

config1['cantNodos'] = config1['Gyms'] + config1['Pokeparadas']
config2['cantNodos'] = config2['Gyms'] + config2['Pokeparadas']
config3['cantNodos'] = config3['Gyms'] + config3['Pokeparadas']
config4['cantNodos'] = config4['Gyms'] + config4['Pokeparadas']


fig, ax = plt.subplots()
ax = config1.groupby('cantNodos').mean()['Tiempo'].plot(color='forestgreen', marker="D", linewidth = 1)
config2.groupby('cantNodos').mean()['Tiempo'].plot(color='purple', marker=">", linewidth = 1, ax=ax)
config3.groupby('cantNodos').mean()['Tiempo'].plot(color='blue', marker="<", linewidth = 1, ax=ax)
config4.groupby('cantNodos').mean()['Tiempo'].plot(color='red', marker="*", linewidth = 1, ax=ax)


#config1['Configuracion:'] = "Ej = 1"
#config2['Configuracion:'] = "Ej = 2"
#config3['Configuracion:'] = "Ej = 3, V = A"
#config4['Configuracion:'] = "Ej = 4, V = A, RCL = 3, Criterio = 2, Limite = 30"


plt.xlabel('n + m')
plt.ylabel('Tiempo en ns')
#ax.set_ylim([-100000000, 1500000000])
plt.rcParams.update({'font.size': 8})

ax.set_yscale('log')

legend = ax.legend(['Ejercicio 1', 'Ejercicio 2', 'Ejercicio 3, vecindad A', 'Ejercicio 4, vecindad A, RCL = 3, criterio 2, límite 30'], loc = 'upper center', bbox_to_anchor = (0.5, -0.1), fancybox = True, shadow = True, ncol = 2)
frame = legend.get_frame()
frame.set_facecolor('white')
frame.set_linewidth(0)


#plt.show()
plt.gcf().set_size_inches(6, 4)
plt.savefig('expAleatOp_cantNodos_tiempo.pdf', format='pdf', bbox_inches = 'tight', dpi = 100)


#result = pd.concat([config1,config2,config3,config4]).reset_index()
#result.plot.line()



#g = sns.plot(x = 'Gyms', y = 'Tiempo', hue = "Configuracion:", ci = 0.68, data = result)
#plt.show()

#sns.barplot(x = 'Gyms', y = 'Distancia', hue = "Configuracion:", ci = 0.68, data = result)
#plt.show()
```

## Distancia vs cantidad de nodos

```python

config1 = ejemplo
#config1 = config1.groupby('RCL')
config2 = ejemplo2
#config2 = config2.groupby('RCL')
config3 = ejemplo3
#config3 = config3.groupby('RCL')
config4 = ejemplo4
#config4 = config4.groupby('RCL')

config1['cantNodos'] = config1['Gyms'] + config1['Pokeparadas']
config2['cantNodos'] = config2['Gyms'] + config2['Pokeparadas']
config3['cantNodos'] = config3['Gyms'] + config3['Pokeparadas']
config4['cantNodos'] = config4['Gyms'] + config4['Pokeparadas']



config1 = config1.groupby('n + m').mean()
#config1 = config1.groupby('RCL')
config2 = config2.groupby('n + m').mean()
#config2 = config2.groupby('RCL')
config3 = config3.groupby('n + m').mean()
#config3 = config3.groupby('RCL')
config4 = config4.groupby('n + m').mean()
#config4 = config4.groupby('RCL')


config1['Configuracion:'] = "Ej = 1"
config2['Configuracion:'] = "Ej = 2"
config3['Configuracion:'] = "Ej = 3, V = A"
config4['Configuracion:'] = "Ej = 4, V = A, RCL = 3, Criterio = 2, Limite = 30"


result = pd.concat([config1,config2,config3,config4])

#g = sns.plot(x = 'Gyms', y = 'Tiempo', hue = "Configuracion:", ci = 0.68, data = result)
#plt.show()

g = sns.barplot(x = 'cantNodos', y = 'Distancia', hue = "Configuracion:", ci = 0.68, data = result)
plt.gcf().set_size_inches(6, 4)
plt.savefig('expAleatOp_cantNodos_distancia.pdf', format='pdf', bbox_inches = 'tight', dpi = 100)
```
